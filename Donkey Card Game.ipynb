```python
def deal_cards_to_players(deck,num_of_players):
    
    #assign player Ids to players
    
    player_id = 1
    player_hand_dict = OrderedDict()
    while(player_id <= num_of_players):
        player_hand = Player_hand(player_id)
        player_hand_dict[player_id] = player_hand
        #player_hand_list.append(player_hand)
        player_id += 1
     
    #distribute cards to players
    #pop out card from main deck and add to player deck
    while(deck.deck_count() > 0):
        #print deck.deck_count()
        for player_hand_key,player_hand_val in player_hand_dict.items():
            if(deck.deck_count() == 0):
                break
            card = deck.deal_card()
            player_hand_val.accept_card(card)
        
    return player_hand_dict
    
```

```python
def get_highest_value_player_id(in_play_deck,in_play_deck_play_order_dict):
    
    sorted_temp_dict = OrderedDict()
    i = 0
    for card in in_play_deck.return_deck():
        play_order = in_play_deck_play_order_dict[i]
        sorted_temp_dict[play_order] = card.get_rank()
        i +=1
        
    maxVal = 0
    maxKey = 0
    for key , val in sorted_temp_dict.items():
        if(int(val)>int(maxVal)):
            maxVal = val
            maxKey = key
        
    if(maxKey>0):
        return maxKey
        
```
